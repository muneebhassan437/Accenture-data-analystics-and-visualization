```python
import csv
import pandas as pd
```

# Content file cleanage

```python
with open ('Content.csv','r') as file:
    reader = csv.reader(file)
    category = []
    blanks = 0
    for row in reader:
        if row[1] == '':
            blanks += 1
        if row[1] != 'Category':
            category.append(row[1])
blanks


        
    
```

```python
for i in range(len(category)):
    category[i] = category[i].lower()
    category[i] = category[i].strip('"')

```

```python
# Step 1: Read the CSV file
df = pd.read_csv('Content.csv')

# Step 3: Update the specific column
df['Category'] = category

# Step 4: Save it back to CSV (overwrite or new file)
df.to_csv('Content_updated.csv', index=False)
```

# Top 5 categories

```python
with open('Output.csv', 'r') as output:
    reader = csv.reader(output)
    top_categories = []
    total_scores = []
    for row in reader:
        if row[0] != 'Category':
            top_categories.append(row[0])
        if row[1] != 'score':
            total_scores.append(row[1])
top_categories
```

```python
new_top_categories = []
for i in range(4,-1,-1):
    new_top_categories.append(top_categories[i])
new_top_categories
```

```python
total_scores.sort()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Plot
plt.figure(figsize=(8, 5))
plt.bar(new_top_categories, total_scores)

# Labels and title
plt.xlabel('Categories')
plt.ylabel('Scores')
plt.title('Top 5 categories')

# Show plot
plt.show()
```

```python
colors = ['yellow', 'green', 'blue', 'purple', 'red']
plt.pie(total_scores, labels=new_top_categories, autopct='%1.1f%%', colors=colors, startangle=90)
plt.title('Top 5 categories')
plt.show()
```
